## 1. What Is a DataFrame?

A DataFrame is essentially a spreadsheet inside Python. It has rows and columns, just like an Excel sheet, but you can manipulate it with code instead of clicking around.

Let's start by importing pandas and building a tiny dataset by hand.

````python
import pandas as pd

# A DataFrame can be created from a dictionary.
# Each key becomes a column name; each value is a list of entries.

data = {
    "country": ["Brazil", "Germany", "Japan", "Nigeria", "United States"],
    "population_millions": [214, 84, 125, 223, 333],
    "gdp_trillions_usd": [1.92, 4.26, 4.23, 0.44, 25.46],
    "region": ["South America", "Europe", "Asia", "Africa", "North America"]
}

df = pd.DataFrame(data)
df
````

That's it — five rows, four columns. The numbers on the left (0, 1, 2, ...) are the **index**. Pandas adds this automatically to give every row a label.

Take a moment to notice the structure: each *row* is a country, and each *column* is a variable we know about that country. This "one row per observation, one column per variable" pattern is the foundation of tidy data, and it's how you'll want to organize most datasets.

## 3. Selecting Data

### Selecting columns

You can grab a single column with bracket notation. This returns a **Series** — pandas' term for a single column of data.

````python
# Single column → returns a Series
df["country"]
````

````python
# Multiple columns → pass a list of column names → returns a DataFrame
df[["country", "gdp_trillions_usd"]]
````

Notice the double brackets `[[ ]]` when selecting multiple columns. The outer brackets are the pandas selection operator; the inner brackets define a Python list.

### Selecting rows

You can select rows by position (like list indexing) using `.iloc`, or by index label using `.loc`.

````python
# .iloc uses integer positions (0-based), just like a Python list
df.iloc[0]       # first row
````

````python
df.iloc[1:3]     # rows at positions 1 and 2 (the end is exclusive, like slicing)
````

````python
# .loc uses the index labels. Since our index is 0, 1, 2, ...
# it looks the same here, but .loc is inclusive on both ends.
df.loc[1:3]      # rows with index labels 1, 2, AND 3
````

The `.iloc` vs `.loc` distinction trips everyone up at first. The short version: use `.iloc` when you're thinking in terms of position ("give me the first three rows"), and `.loc` when you're thinking in terms of labels.

## 4. Filtering Rows

This is where pandas starts to feel powerful. You can filter a DataFrame by writing a condition, and pandas returns only the rows where that condition is `True`.

Let's say we want countries with GDP above $4 trillion.

````python
# Step 1: The condition itself produces a Series of True/False values
df["gdp_trillions_usd"] > 4
````

````python
# Step 2: Pass that True/False series back into the DataFrame
# Pandas keeps only the rows where the value is True

high_gdp = df[df["gdp_trillions_usd"] > 4]
high_gdp
````

You can combine conditions with `&` (and) and `|` (or). Each condition needs its own parentheses.

````python
# Countries with GDP > 4 trillion AND population > 100 million
df[(df["gdp_trillions_usd"] > 4) & (df["population_millions"] > 100)]
````

## 5. Adding and Modifying Columns

You can create new columns from existing ones, just like adding a formula column in a spreadsheet.

````python
# GDP per capita (in USD)
# GDP is in trillions, population is in millions,
# so (trillions / millions) = millions → multiply by 1e6 to get USD

df["gdp_per_capita"] = (df["gdp_trillions_usd"] / df["population_millions"]) * 1e6
df
````

Notice how the arithmetic operated on every row at once — no `for` loop needed. This is one of pandas' key advantages: **vectorized operations** let you work on entire columns in a single expression.

## 9. Handling Messy Data

Real data is messy. Numbers get stored as text, cells are empty, rows contain junk. Pandas gives you tools to deal with this.

### Converting types with `pd.to_numeric()`

````python
# Sometimes numbers arrive as strings (this is common with Excel data)
messy = pd.DataFrame({
    "country": ["Chile", "Egypt", "Poland", "Somalia"],
    "inflation": ["4.5", "13.9", "N/A", "6.1"]   # note: these are strings, not numbers
})

messy.dtypes
````

````python
# pd.to_numeric() converts a column to numbers.
# errors='coerce' turns anything it can't convert (like "N/A") into NaN
# (NaN = "Not a Number" — pandas' way of representing missing data)

messy["inflation"] = pd.to_numeric(messy["inflation"], errors="coerce")
messy
````

````python
# Now the dtype is float64 — a real number
messy.dtypes
````

### Checking for missing values

````python
# .isna() returns True where data is missing
messy["inflation"].isna()
````

````python
# .notna() is the opposite — True where data exists
# This is handy for filtering out missing values
messy[messy["inflation"].notna()]
````

## 10. Using `.apply()` and Lambda Functions

Sometimes you need to apply a custom operation to every value in a column. The `.apply()` method does exactly this — it runs a function on each element (or each row) and returns the results.

A **lambda function** is just a small, anonymous function you can write inline instead of defining with `def`.

````python
# Quick refresher on lambda:
# A regular function:
def double(x):
    return x * 2

# The same thing as a lambda:
double_lambda = lambda x: x * 2

print(double(5))         # 10
print(double_lambda(5))  # 10
````

````python
# Using .apply() with a lambda on our panel data
# Let's classify each inflation observation as "high" or "normal"

panel["category"] = panel["inflation_pct"].apply(lambda x: "high" if x > 10 else "normal")
panel
````

You can also use `.apply()` to filter. Instead of the vectorized comparison we saw earlier (`df[df["col"] > value]`), you can write:

```python
df[df["col"].apply(lambda x: x > value)]
```

Both produce the same result. The lambda version is more flexible — you can put any logic you want inside it — but the vectorized version is faster on large datasets. Part II asks you to practice with lambda, so you'll use `.apply()` there.

````python
# Filtering with lambda: same result as panel[panel["inflation_pct"] > 10]
panel[panel["inflation_pct"].apply(lambda x: x > 10)]
````

## 11. Grouping and Aggregating

One of the most common operations in data analysis is splitting your data into groups, computing something for each group, and combining the results. Pandas calls this **groupby**.

Example: What's the average inflation rate for each country across all years?

````python
panel.groupby("country")["inflation_pct"].mean()
````

That one line does a lot:
1. **Groups** the rows by country
2. **Selects** the inflation column within each group
3. **Computes** the mean for each group

You can chain `.sort_values()` to rank the results, and `.head(n)` to get the top N.

````python
# Top 2 countries by average inflation
panel.groupby("country")["inflation_pct"].mean().sort_values(ascending=False).head(2)
````

### Turning a grouped result back into a DataFrame

The result of `.groupby().mean()` is a **Series** (one column, indexed by the group). If you want it as a proper DataFrame with columns, use `.reset_index()`.

````python
avg_inflation = panel.groupby("country")["inflation_pct"].mean().sort_values(ascending=False)

# Before reset_index — it's a Series with 'country' as the index
print(type(avg_inflation))
print(avg_inflation)
print()
````

````python
# After reset_index — it's a DataFrame with 'country' as a regular column
avg_inflation_df = avg_inflation.reset_index()
print(type(avg_inflation_df))
avg_inflation_df
````

## 13. Merging DataFrames

Sometimes the data you need lives in two separate tables and you need to join them. Pandas' `pd.merge()` works like a database join or an Excel VLOOKUP.

````python
# Suppose we have a separate table of crisis indicators
crisis_data = pd.DataFrame({
    "country": ["Argentina", "Argentina", "Germany"],
    "year": [2020, 2022, 2022],
    "crisis_type": ["currency", "currency", "energy"]
})

crisis_data
````

````python
# Merge on shared columns (country AND year)
# how="left" keeps all rows from the left DataFrame,
# filling NaN where there's no match in the right DataFrame.

merged = pd.merge(panel, crisis_data, on=["country", "year"], how="left")
merged
````

The NaN values in `crisis_type` mean "no crisis recorded for that country-year." You could now filter for rows where a crisis occurred using `.notna()`.

````python
merged[merged["crisis_type"].notna()]
````
