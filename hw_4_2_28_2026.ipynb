Calculate Trade Balance

Exports: 500 million dollars
Imports: 600 million dollars

Task: Define a function trade_balance(exports, imports) and use it to determine the trade balance.

```python
#Your code here:
def trade_balance(exports, imports):
    if exports - imports > 0:
        return "surplus"
    elif exports - imports == 0:
        return "balanced trade - woah"
    else:
        return "deficit"

trade_balance(20, 1)
```

Summarize Human Development Index (HDI)
Write a function that categorizes a country’s Human Development Index (HDI) score into "very high", "high", "medium", or "low" based on the following ranges:

Very high: HDI ≥ 0.8
High: 0.7 ≤ HDI < 0.8
Medium: 0.55 ≤ HDI < 0.7
Low: HDI < 0.55
Use the following data:

Norway HDI: 0.953
Task: Define a function hdi_summary(country, hdi) that prints the category for a given country and HDI score.

Find Largest Economy
Given the GDPs of three countries, write a function that determines which country has the largest economy. Use the following GDP data (in billion dollars):

USA: 21,000
China: 14,000

Germany: 4,200
Task: Define a function largest_economy(gdp_data) that takes a dictionary with country names as keys and GDP as values, then prints the largest economy.


```python
#Your code here:
def largest_economy(gdp_data):

    max_nation = None
    max_value = 0
    for nation, value in gdp_data.items():
        if value > max_value:
            max_value = value
            max_nation = nation

    return f"{max_nation} has the largest economy with a GDP of {max_value} billion dollars"

largest_economy({"USA": 21000,
        "China": 14000,
        "Germany": 4200})
```

Sanction Impact on Economy
A country’s economy shrinks after sanctions are imposed. Write a function to calculate the percentage reduction in GDP. Use the following data:

Original GDP: 1000 billion dollars
GDP after sanctions: 850 billion dollars

Task: Define a function sanction_impact(original_gdp, post_sanction_gdp) and use it to calculate the impact of sanctions.


```python
#Your code here:
def sanction_impact(original_gdp, post_sanction_gdp):
    percent = post_sanction_gdp / original_gdp - 1
    return f"{percent * 100}% change in GDP"

sanction_impact(1000, 850)
```

### Passing a function to another function. We can pass the return value of a function to another function.

A country’s diplomatic influence is determined by the size of its economy (GDP in billions) and its military spending (in billions).

First, write a function called calculate_influence(economy_size, military_spending) that returns the total influence score by adding the two values together. 

Then, write a second function called compare_influence(influence1, influence2) that takes two influence scores and prints which country has greater influence, or prints that they are equal if the scores match. 

You must call calculate_influence() for two different countries and pass the returned values into compare_influence() for comparison.

You must test your functions using the following three scenarios:

1.Country A Has Greater Influence

Country A: GDP = 800, Military = 200

Country B: GDP = 700, Military = 150

Expected result: Country A has greater influence.

2. Country B Has Greater Influence

Country A: GDP = 600, Military = 150

Country B: GDP = 700, Military = 250

Expected result: Country B has greater influence.

3. Equal Influence

Country A: GDP = 500, Military = 200

Country B: GDP = 600, Military = 100

Expected result: Both countries have equal influence.


```python
#Your code here:
def calculate_influence(economy_size, military_spending):
    total_influence = economy_size + military_spending
    return total_influence

def compare_influence(influence1,influence2):
    if calculate_influence(*influence1) > calculate_influence(*influence2):
        return "Country A has greater influence"
    elif calculate_influence(*influence1) == calculate_influence(*influence2):
        return "Both countries have equal influence"
    else:
        return "Country B has greater influence"

compare_influence((500,200),(600,100))
```

## Translating Sentences:

"Make a career of humanity. Commit yourself to the noble struggle for equal rights. You will make a better person of yourself, a greater nation of your country, and a finer world to live in."


```python
#Your code here:
from deep_translator import GoogleTranslator

def passage(entry, target_language):
    return GoogleTranslator(source="auto", target=target_language).translate(entry)

passage("Make a career of humanity. Commit yourself to the noble struggle for equal rights. You will make a better person of yourself, a greater nation of your country, and a finer world to live in.", "chinese (simplified)")
```

```python
menu = {
    "Pizza": {
        "Basic": "Tomato and Cashew Cream Cheese",
        "Size": {
            "Small": 6.95,
            "Medium": 8.95,
            "Large": 10.90
        },
        "Toppings": {
            "Shitake": 0.50,
            "Jackfruit": 0.50,
            "Sweet Potato": 0.50,
            "Hummus": 0.50,
            "Chickpeas": 0.50,
            "Lentils": 0.50,
            "Tofu": 0.50,
            "Edamame": 0.50,
            "Black Bean": 0.50,
            "Red Cabbage": 0.50,
            "Cucumber": 0.50,
            "Tomato": 0.50,
            "Jicama": 0.50,
            "Fried Green Tomatoes": 0.50,
            "Sun Dried Tomatoes": 0.50
        }
    },
    "Salad": {
        "Size": {
            "Small": 4.95,
            "Large": 6.95
        },
        "Type": ["Vegan Caesar", "Vegan Greek", "Cucumber_Tumeric_Tomato"]
    }
}

meat = ["pepperoni", "sausage", "ham", "normal cheese", "mozzarella", "parmesan"]
done_signal = ["nothing", "that's it", "nope", "no, that's it"]

last_rejected_name = ""

while True:
    phone = input("What number would you like to call?: ")

    while phone not in ["203-555-5555", "2035555555", "(203) 555-5555"]:
        print("Number not found")
        phone = input("What number would you like to call?: ")

    name = input("Elm City Pizza, where we have never had a case of food poisoning that can be proved. My name is Hal, what's your name? ")

    if name == last_rejected_name:
        print(f"Sorry {name} but that's too quick to evolve. Call us back when you have.")
        continue

    order = input(f"Wow {name} what an awesome name. What would you like to order? ").lower()

    while order not in ["pizza", "salad"]:
        print(f"Sorry {name}, we don't have that.")
        print(menu)
        order = input("How about a super awesome pizza? Take a look at our menu, you'll find it's delicious. ").lower()

    cust_order = []
    total_cost = 0
    restart_call = False

    if order == "pizza":
        print(menu["Pizza"])
        cust_order.append("Pizza")

        topping = input("What toppings do you want? ").lower()

        while True:
            if topping in done_signal:
                break

            elif topping in meat:
                print(f"Sorry, {name}, but you need to do some more work evolving.")
                last_rejected_name = name
                restart_call = True
                break

            elif topping in [x.lower() for x in menu["Pizza"]["Toppings"].keys()]:
                cust_order.append(topping)
                total_cost += 0.50
                topping = input("Anything else? ").lower()

            else:
                topping = input(f"I'm afraid we don't have that, Professor King. Oops - I mean {name}. Anything else? ").lower()

        if restart_call:
            continue

        size = input("What size? Small, Medium or Large? ").title()

        while size not in menu["Pizza"]["Size"]:
            size = input(f"I'm afraid we don't have that, Professor King. Oops - I mean {name}. Small, Medium or Large? ").title()

        cust_order.append(size)
        total_cost += menu["Pizza"]["Size"][size]

    elif order == "salad":
        print(menu["Salad"])
        cust_order.append("Salad")

        salad = input("What salad do you want? ").title()

        while salad not in menu["Salad"]["Type"]:
            salad = input(f"I'm sorry, we don't have that, Professor King. Oops - I mean {name}. Check the menu above for our options. ").title()

        cust_order.append(salad)

        size = input("What size? Small or Large? ").title()

        while size not in menu["Salad"]["Size"]:
            size = input(f"I'm afraid we don't have that, Professor King. Oops - I mean {name}. Small or Large? ").title()

        cust_order.append(size)
        total_cost += menu["Salad"]["Size"][size]

    print("Your order is:", cust_order)
    print("Your total is: $", round(total_cost, 2))
    break
```
