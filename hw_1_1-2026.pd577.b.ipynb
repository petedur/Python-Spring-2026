15. Write a script that takes a user inputted target_string of your chosing and prints it out in the following three formats.

All letters capitalized - capitalize( )
All letters lowercase - lower( )
All vowels lowercase and all consonants uppercase 

```python
target_script = input("enter a string: ")
print("All letters capitalized - ", target_script.upper())
print("All letters lowercase - ", target_script.lower())
vowels = "aeiou"
final = "".join(
    ch.lower() if ch.lower() in vowels
    else ch.upper() if ch.isalpha()
    else ch
    for ch in target_script
)
print("All vowels lowercase and all consonants uppercase - "
      , final)
```
