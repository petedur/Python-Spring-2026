````python
# TODO: write your class here
class Book:
    def __init__(self, title, author, pages, read=False, genre="Unknown", dwd_id=None):
        self.title = title
        self.author = author
        self.pages = pages
        self.read = read
        self.genre = genre
        self.dwd_id = dwd_id

    def describe(self):
        print(f'"{self.title}" is a {self.genre} book by {self.author}. It has {self.pages} pages and I have {'had a chance to' if self.read else 'not had a chance to'} read the book.')

    def toggle(self):
        self.read = not self.read
        print(f'I have {'now had a chance to' if self.read else 'still not had a chance to'} read "{self.title}."')

    def update(self, new_value):
        if new_value <= 0:
            raise ValueError("Quit fooling around, silly goose. Pages must be greater than 0.")
    
        old_pages = self.pages
        self.pages = new_value
        print(f"Oops, it actually has {self.pages} pages! Updated pages from {old_pages} to {self.pages}.")

    def compare(self, other):
        if self.pages > other.pages:
            print(f'"{self.title}" has more pages than "{other.title}."')
        elif self.pages < other.pages:
            print(f'"{self.title}" has fewer pages than "{other.title}."')
        else:
            print(f'Twinsies! "{self.title}" and "{other.title}" have the EXACT same number of pages!')

    def well_read_guess(self):
        if self.read and self.pages > 100 and self.genre == "classic":
            print("Wow, I'm guessing you are pretty well read! And handsome, too.")
        elif self.read and self.genre != classic:
            print("Cool book, but your english teacher has some recommendations. At least you're handsome!")
        else:
            print("You're handsome, but you should read more.")
````

> ### 📘 New Concept — `getattr(obj, name)`
>
> So far you have accessed attributes directly: `obj.speed`, `obj.name`, etc.
> That works fine when you know the attribute name at the time you write the code.
> But what if the name is stored in a variable — chosen at runtime by the user?
>
> `getattr()` solves this. It takes an object and a **string** containing the
> attribute name, and returns that attribute's value:
>
> ```python
> # These two lines do exactly the same thing:
> value = obj.speed          # attribute name is hardcoded
> value = getattr(obj, 'speed')  # attribute name is a string
> ```
>
> Because the name is just a string, you can pass it in as a variable:
>
> ```python
> attribute = input('Which attribute? ')   # e.g. user types 'speed'
> value = getattr(obj, attribute)          # looks up obj.speed at runtime
> ```
>
> If the attribute does not exist, `getattr()` raises an `AttributeError` —
> which you can catch with a `try-except`, as you will do in the question below.
>
> You can also supply a **fallback default** as a third argument, so no error is raised:
> ```python
> value = getattr(obj, attribute, None)  # returns None if attribute doesn't exist
> ```

## Question 4 — Searching & Filtering ⭐⭐⭐⭐

A registry is only useful if you can find things in it quickly. In this question you will write two search tools.

---

### Tool 1 — `search_by_attribute(registry, attribute, value)`

- Loops through every object in the registry.
- Collects all objects where `getattr(obj, attribute)` equals `value` — use `getattr()`, do not hardcode the attribute name.
- Returns a **list** of matching objects.
- If **no matches** are found → print an informative message and return an empty list.
- If the **attribute does not exist** on the objects → catch the `AttributeError` and print a helpful message.

---

### Tool 2 — `filter_by_range(registry, attribute, min_val, max_val)`

- Works only on **numeric** attributes.
- Returns a list of objects where the attribute's value falls within `[min_val, max_val]` **inclusive**.
- Raises a `TypeError` with a clear message if the attribute's value is not a number.
- If the result list is **empty** → print a message saying no objects fell within that range.

---

### Demonstrate both tools:
- Call `search_by_attribute` with at least **two different attributes**.
- Call `filter_by_range` including one test where **the range contains no results**.
- For each result list, print the results using your `describe` method.

````python
# TODO: write search_by_attribute here
def search_by_attribute(registry, attribute, value):
    matches = []

    try:
        for obj in registry.values():
            if getattr(obj, attribute) == value:
                matches.append(obj)

    except AttributeError:
        print(f'Sorry handsome, "{attribute}" is NOT a valid attribute.')
        return []

    if not matches:
        print(f'Sorry handsome, no matches found for {attribute} = {value}.')
        return []

    return matches
````

````python
# TODO: write filter_by_range here
def filter_by_range(registry, attribute, min_val, max_val):
    matches = []

    try:
        for obj in registry.values():
            value = getattr(obj, attribute)

            if not isinstance(value, (int, float)):
                raise TypeError(f'"{attribute}" must be numeric to use filter_by_range.')

            if min_val <= value <= max_val:
                matches.append(obj)

    except AttributeError:
        print(f'Sorry handsome, "{attribute}" is NOT a valid attribute.')
        return []

    except TypeError as e:
        print(e)
        return []

    if not matches:
        print(f'Sorry handsome, no matches found for {attribute} between {min_val} and {max_val}.')
        return []

    return matches
````

> ### 📘 New Concept — Subclasses & Inheritance
>
> A **subclass** is a class that is built on top of an existing one.
> The existing class is called the **parent** (or base) class; the new one is the **child** (or subclass).
> The child automatically inherits everything the parent has — all its attributes and methods —
> without you having to rewrite any of it.
>
> You create a subclass by passing the parent class as an argument:
>
> ```python
> class Animal:                      # parent class
>     def __init__(self, name, age):
>         self.name = name
>         self.age  = age
>
>     def describe(self):
>         print(f'{self.name}, age {self.age}')
>
> class Dog(Animal):                 # Dog is a subclass of Animal
>     def __init__(self, name, age, breed):
>         super().__init__(name, age) # run the parent's __init__ first
>         self.breed = breed          # then add what's new
>
>     def fetch(self):               # a method only Dogs have
>         print(f'{self.name} fetches the ball!')
> ```
>
> **`super()`** is how the child talks to its parent. Calling `super().__init__()`
> runs the parent's setup code so you don't have to copy it into the child.
>
> The child can also **override** a parent method by defining one with the same name.
> Python will always use the child's version when the method is called on a child object:
>
> ```python
>     def describe(self):                      # overrides Animal.describe
>         super().describe()                   # optionally call the parent's version first
>         print(f'Breed: {self.breed}')        # then add the extra detail
> ```
>
> This means a `Dog` object has access to everything `Animal` defines,
> plus anything extra added in `Dog` itself. A plain `Animal`, however,
> knows nothing about `Dog` — inheritance only flows **downward**.
>
> When you loop over a collection containing both `Animal` and `Dog` objects
> and call `describe()` on each one, Python automatically calls the right version
> for each object. This behaviour is called **polymorphism**, and it is what
> Question 5 Part C asks you to observe.

````python
# TODO: write each menu handler function here
def run_manager(registry):
    while True:
        print("\n #**Registry Manager**#")
        print("[1] Add a new entry")
        print("[2] View all entries")
        print("[3] Search by attribute")
        print("[4] Filter by numeric range")
        print("[5] Edit an entry")
        print("[6] Remove an entry")
        print("[7] Show statistics")
        print("[0] Quit")
        try:
            choice = input("Choose an option: ").strip()
            if choice == "1":
                handle_add(registry)
            elif choice == "2":
                handle_view(registry)
            elif choice == "3":
                handle_search(registry)
            elif choice == "4":
                handle_filter(registry)
            elif choice == "5":
                handle_edit(registry)
            elif choice == "6":
                handle_remove(registry)
            elif choice == "7":
                handle_stats(registry)
            elif choice == "0":
                print("Goodbye!")
                break
            else:
                print("Invalid menu choice. Please try again.")
        except Exception as e:
            print(f"Something went wrong: {e}")

    # def handle_add(registry): ...
def handle_add(registry):
    try:
        type = input("Add [1] Book or [2] EBook? ").strip()
        if type == "1":
            obj = create_book_from_input()
        elif type == "2":
            obj = create_ebook_from_input()
        else:
            print("Silly goose! Hit [1] or [2]!!!!")
            return
        add_to_registry(registry, obj)
    except Exception as e:
        print(f"Error adding entry: {e}")
    
# def handle_view(registry): ...
def handle_view(registry):
    if not registry:
        print("The registry is empty - we've been robbed!")
        return

    print(f"Total entries: {len(registry)}")
    for i, obj in enumerate(registry.values(), start=1):
        print(f"Entry {i}:")
        obj.describe()
        
# def handle_search(registry): ...
def handle_search(registry):
    try:
        type = input("Search by [1] title, [2] author, [3] number of pages, [4] dewey decimal system ID, [5] read, [6] genre, [7] file size, or [8] device type?").strip()
        if type == "1":
            def handle_search_by_title(registry):
                title = input("Enter a book title: ").strip()
                results = search_by_attribute(registry, "title", title)

                for book in results:
                    book.describe()
        elif type == "2":
            def handle_search_by_author(registry):
                title = input("Enter a book author: ").strip()
                results = search_by_attribute(registry, "author", author)

                for book in results:
                    book.describe()
        elif type == "3":
            def handle_search_by_pages(registry):
                title = input("Enter a number of pages: ").strip()
                results = search_by_attribute(registry, "pages", pages)

                for book in results:
                    book.describe()
        elif type == "4":
            def handle_search_by_dwd_id(registry):
                title = input("Enter a dewey decimal system ID: ").strip()
                results = search_by_attribute(registry, "dwd_id", dwd_id)

                for book in results:
                    book.describe()
        elif type == "5":
            def handle_search_by_read(registry):
                title = input("Enter true or false: ").strip()
                results = search_by_attribute(registry, "read", read)

                for book in results:
                    book.describe()
        elif type == "6":
            def handle_search_by_read(registry):
                title = input("Enter a genre: ").strip()
                results = search_by_attribute(registry, "genre", genre)

                for book in results:
                    book.describe()
        elif type == "7":
            def handle_search_by_file_size(registry):
                title = input("Enter a file size: ").strip()
                results = search_by_attribute(registry, "file_size", file_size)

                for book in results:
                    book.describe()
        elif type == "8":
            def handle_search_by_device_type(registry):
                title = input("Enter a device type: ").strip()
                results = search_by_attribute(registry, "device_type", device_type)

                for book in results:
                    book.describe()
        else:
            print("Silly goose! Hit one of the numbers from the menu!")
            return
        add_to_registry(registry, obj)
    except Exception as e:
        print(f"Error adding entry: {e}")

# def handle_filter(registry): ...
def filter_by_range(registry, attribute, min_val, max_val):
    matches = []

    try:
        for obj in registry.values():
            value = getattr(obj, attribute)

            if not isinstance(value, (int, float)):
                raise TypeError(f'"{attribute}" must be numeric to use filter_by_range.')

            if min_val <= value <= max_val:
                matches.append(obj)

    except AttributeError:
        print(f'Sorry handsome, "{attribute}" is NOT a valid attribute.')
        return []

    except TypeError as e:
        print(e)
        return []

    if not matches:
        print(f'Sorry handsome, no matches found for {attribute} between {min_val} and {max_val}.')
        return []

    return matches

# def handle_edit(registry): ...
def handle_edit(registry, key, updated_object):
    if key not in registry:
        print(f'Sorry mate, no dice! "{key}" is not found. Unfortunately, nothing updated')
        return

    registry[key].describe()
        
    overwrite_confirmation = input(f'Ayo! An entry for "{key}" already exists. Are you sure you want to overwrite it? Please say yes or no.').lower().strip()
    
    if overwrite_confirmation != "yes":
        print("This right here is a cancellation message. Cancellation I say!")
        return
    
    try:
        registry[key] = updated_object
        print(f'Made an update to the entry with key "{key}".')
        registry[key].describe()
    except Exception as e:
        print(f"Ruh roh, something went wrong during the update: {e}")
# def handle_remove(registry): ...
def remove_from_registry(registry, key):
    if key not in registry:
        print(f'Good news - this is already not found in the registry! Nothing removed.')
        return

    registry[key].describe()

    confirmation = input(
        f'Ayo! An entry for "{key}" exists. Are you sure you want to delete it? Please say yes or no. '
    ).lower().strip()

    if confirmation != "yes":
        print("Deletion cancelled.")
        return

    try:
        del registry[key]
        print(f'Entry with key "{key}" was removed successfully.')
    except Exception as e:
        print(f'Ruh roh, something went wrong during removal: {e}')

# def handle_stats(registry): ...
def handle_stats(registry):
    total_entries = len(objects)
    highest = max(objects, key=lambda obj: obj.pages)
    lowest = min(objects, key=lambda obj: obj.pages)
    average_pages = sum(obj.pages for obj in objects) / total_entries
    read_true = sum(1 for obj in objects if obj.read)
    read_false = total_entries - read_true
    print(f"Total entries: {total_entries}")
    print("Most pages page count:")
    highest.describe()
    print("Fewest pages page count:")
    lowest.describe()
    print(f"Average pages: {average_pages:.2f}")
    print(f"Read = True: {read_true}")
    print(f"Read = False: {read_false}")
````
