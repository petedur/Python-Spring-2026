## Question 4 - Medium: social incentives data (20 points)

In your policy lab, you are studying whether places with higher trust and fairness might be more willing to support common projects.

Run the starter cell below. Then:

1. Write a function `solidarity_score(trust, fairness, stress)` that returns:

   `trust + fairness - stress`

2. Use a `for` loop to build a new dictionary called `scores` where each city is matched with its solidarity score.
3. Use `if` / `elif` / `else` to classify each city:
   - if the score is greater than or equal to 100, label it `"high solidarity"`
   - if the score is greater than or equal to 70, label it `"medium solidarity"`
   - if the score is less than 70, label it `"low solidarity"`
4. Store these labels in a dictionary called `categories`.
5. Use `sorted()` to create a leaderboard from highest score to lowest score.
6. Print:
   - the `scores` dictionary,
   - the `categories` dictionary,
   - the sorted leaderboard,
   - and the city with the highest solidarity score.

This question intentionally turns one of our course tangents into structured data analysis.


````python
social_incentives = {
    "New Haven": {"trust": 68, "fairness": 61, "stress": 35},
    "Detroit": {"trust": 52, "fairness": 44, "stress": 49},
    "Austin": {"trust": 63, "fairness": 58, "stress": 28},
    "Fresno": {"trust": 47, "fairness": 40, "stress": 55},
    "Madison": {"trust": 74, "fairness": 70, "stress": 22}
}
````

````python
# Write your code for Question 4 here
def solidarity_score(trust, fairness, stress):
    return trust + fairness - stress

scores = {}
categories = {}

for city in social_incentives:
    trust = social_incentives[city]["trust"]
    fairness = social_incentives[city]["fairness"]
    stress = social_incentives[city]["stress"]

    score = solidarity_score(trust, fairness, stress)
    scores[city] = score

    if score >= 100:
        categories[city] = "high solidarity"
    elif score >= 70:
        categories[city] = "medium solidarity"
    else:
        categories[city] = "low solidarity"

cities_sorted = sorted(scores)

print("1. Scores Dictionary: ", scores)
print("2. Categories Dictionary:", categories)
print("3. Sorted Leaderboard:", cities_sorted)
print("4. City with the highest solidarity score:", cities_sorted[0])
````

4. Run 1000 simulations and store the number of changes from each simulation in a list called `simulated_changes`.
5. Compute the average of `simulated_changes`.
6. Use `if` / `elif` / `else` to print a short conclusion:
   - if `actual_changes` is smaller than the average simulated value, say the real history looks **more stable**
   - if `actual_changes` is larger than the average simulated value, say the real history looks **more random**
   - otherwise say they look **about the same**

This question intentionally turns another course tangent into a simple simulation exercise.


````python
random.seed(42)

history = [
    "Uruguay",
    "Germany",
    "Brazil",
    "Brazil",
    "England",
    "Brazil",
    "Germany",
    "Argentina",
    "Italy",
    "Argentina",
    "Germany",
    "Brazil",
    "France",
    "Brazil",
    "Italy",
    "Spain",
    "Germany",
    "France",
    "Argentina"
] #these are the last 75 years of world cup winners in order

teams = [
    "Argentina", "France", "Brazil", "Germany", "Spain", "Italy", "England", "Uruguay",
    "Mexico", "Belgium", "Netherlands", "Switzerland", "Sweden", "Russia", "South Korea",
    "United States", "Chile", "Paraguay", "Poland", "Hungary", "Portugal", "Croatia",
    "Denmark", "Austria", "Czech Republic", "Romania", "Scotland", "Turkey", "Serbia",
    "Japan", "Morocco", "Cameroon", "Nigeria", "Tunisia", "Saudi Arabia", "Iran",
    "Australia", "Costa Rica", "Colombia", "Peru", "Ecuador", "Bolivia", "Algeria",
    "Ghana", "Senegal", "Ivory Coast", "South Africa", "Egypt", "United Arab Emirates",
    "Iraq", "Qatar", "Canada", "Honduras", "El Salvador", "Panama", "Jamaica",
    "Cuba", "Haiti", "Trinidad and Tobago", "China", "North Korea", "Israel", "Kuwait",
    "Indonesia", "India", "DR Congo", "Togo", "Angola", "Ukraine",
    "Greece", "Norway", "Wales", "Northern Ireland", "Slovakia", "Slovenia",
    "Bosnia and Herzegovina", "New Zealand", "Iceland", "Czechoslovakia"
] #these are the 80 teams that have qualified for the world cup over its history
````

````python
# Write your code for Question 5 here

def count_changes(history):
    changes = 0
    for i in range(len(history)-1):
        if history[i] != history[i + 1]:
            changes += 1
    return changes

print("PART 1: There have been", count_changes(history), "changes in winner over the last 75 years of world cups")

def simulate_season(teams, tournaments):
    season = []
    for i in range(tournaments):
        season.append(random.choice(teams))
    return season

tournaments = 1000
sim_history = simulate_season(teams, tournaments)
print("PART 2: List of random world cup winners", sim_history)
print("PART 3: Out of", tournaments, "simulated world cups, there have been", count_changes(sim_history), "changes in winner.")

if count_changes(sim_history)/tournaments > count_changes(history)/(len(history)-1):
    print("PART 4: Real history looks more stable.")
elif count_changes(sim_history)/tournaments < count_changes(history)/(len(history)-1):
    print("PART 4: Real history looks more random.")
else:
    print("PART 4: Real and sim history look the same.")
````
